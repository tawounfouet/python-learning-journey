## ⚙️ Méthodes spéciales
```python
__init__ (constructeur) : Initialise l’objet.

__str__(affichage utilisateur)

__repr__ (debug)
```  

````python
def __init__(self, nom, age):
    self.nom = nom
    self.age = age
````

````python
def __str__(self):
    return f"{self.nom} ({self.age} ans)"
````

````python
def __repr__(self):
    return f"Personne(nom={self.nom}, age={self.age}"
````

````python
# Mise en pratique

class Personne:
    def __init__(self, nom, age):
        self.nom = nom
        self.age = age 

    def __str__(self):
        return f"{self.nom} ({self.age} ans)"

    def __repr__(self):
        return f"Personne(nom={self.nom}, age={self.age})"

Personne.__dict__
````

````python
# _init__ (constructeur) : Initialise l’objet
p1 = Personne("Thomas", 30)

p1.__dict__
````

````python
# __repr__ (debug)
p1 
````

````python
# __str__(affichage utilisateur)
print(p1)
````

## 🔷 2.2 Héritage et polymorphisme
### 🧬 Héritage

````python
class Animal:
    def parler(self):
        print("L'animal fait un bruit")

class Chien(Animal):
    def parler(self):
        print("Le chien aboie")
````

### 🔁 Polymorphisme (même méthode, comportement différent)

````python
animaux = [Animal(), Chien()]

for a in animaux:
    a.parler()
````

### 🧠 Méthodes statiques et de classe

````python
class MathUtils:

    @staticmethod
    def addition(a, b):
        return a + b

    @classmethod
    def description(cls):
        return f"Classe: {cls.__name__}"
````

## 🔷 2.3 Encapsulation & abstraction
### 🔒 Attributs privés

````python
class Compte:
    def __init__(self, solde):
        self.__solde = solde  # privé

Compte.__dict__
````

### 🎛️ Propriétés avec @property

````python
class Compte:
    def __init__(self, solde):
        self.__solde = solde

    @property
    def solde(self):
        return self.__solde

    @solde.setter
    def solde(self, valeur):
        if valeur >= 0:
            self.__solde = valeur


Compte.__dict__
````

### ✅ 1. Classe Rectangle

````python
class Rectangle:
    def __init__(self, largeur, hauteur):
        self.largeur = largeur
        self.hauteur = hauteur

    def aire(self):
        return self.largeur * self.hauteur

    def perimetre(self):
        return 2 * (self.largeur + self.hauteur)

# Test
r = Rectangle(5, 3)
print("Aire:", r.aire())
print("Périmètre:", r.perimetre())
````

### ✅ 2. Classe Voiture + héritage

````python
class Voiture:
    def __init__(self, marque, vitesse):
        self.marque = marque
        self.vitesse = vitesse

    def rouler(self):
        return f"{self.marque} roule à {self.vitesse} km/h"


class VoitureElectrique(Voiture):
    def __init__(self, marque, vitesse, batterie):
        super().__init__(marque, vitesse)
        self.batterie = batterie

    def rouler(self):
        return f"{self.marque} électrique roule silencieusement à {self.vitesse} km/h"


class VoitureEssence(Voiture):
    def rouler(self):
        return f"{self.marque} thermique fait du bruit à {self.vitesse} km/h"


# Test
v1 = VoitureElectrique("Tesla", 120, 80)
v2 = VoitureEssence("BMW", 150)

print(v1.rouler())
print(v2.rouler())
````

### ✅ 3. Système de bibliothèque

````python
class Bibliotheque:
    def __init__(self):
        self.livres = []

    def ajouter_livre(self, titre):
        self.livres.append(titre)

    def supprimer_livre(self, titre):
        if titre in self.livres:
            self.livres.remove(titre)

    def rechercher_livre(self, titre):
        return titre in self.livres


Bibliotheque.__dict__
````

````python
# Test
biblio = Bibliotheque()
biblio.livres
````

````python
biblio.ajouter_livre("Python pour débutants")
biblio.ajouter_livre("Django avancé")

biblio.livres
````

````python
print(biblio.rechercher_livre("Python pour débutants"))  # True
````

````python
biblio.supprimer_livre("Python pour débutants")
print(biblio.rechercher_livre("Python pour débutants"))  # False
````

## 🔧 Étape 4 — Ajouter une vraie logique métier

````python
print(r.aire())
````

## 🧠 Étape 5 — Introduire __init__ (le vrai tournant)

````python
class Rectangle:
    def __init__(self, largeur, hauteur):
        self.largeur = largeur
        self.haureur = hauteur

    def aire(self):
        return self.largeur * self.hauteur

Rectangle.__dict__
````

👉 Différence clé :
- Avant : objet vide puis configuré
- Maintenant : objet valide dès sa création

````python
r = Rectangle(6, 5)
r.__dict__
````

````python
r.aire()
````

### 🔥 Différence fondamentale
**❌ Sans __init__**
- objet vide
- risque d’erreur
- dépend d’un ordre d’appel

**✅ Avec __init__**
- objet valide dès la création
- plus sûr
- plus lisible



## 🔒 Étape 6 — Amélioration (robustesse)

````python
class Rectangle:
    def __init__(self, largeur, hauteur):
        if largeur <= 0 or hauteur <= 0:
            raise ValueError("Dimension invalidés")
        self.largeur = largeur
        self.hauteur = hauteur

Rectangle.__dict__
````

````python
r = Rectangle(-2, 5)
````

````python
r = Rectangle(2, 5)
r.__dict__
````

````python
class Utilisateur:
    def __init__(self, nom, email):
        self.nom = nom
        self.email = email

    def afficher_profil(self):
        return f"{self.nom} - {self.email}"


Utilisateur.__dict__
````

````python
# ✅ CORRECTION (niveau propre)

class Utilisateur:
    def __init__(self, nom, email):
        if not nom:
            raise ValueError("Le nom est obligatoire")

        if "@" not in email:
            raise ValueError("Email invalide")

        self.nom = nom
        self.email = email

    def afficher_profil(self):
        return f"{self.nom} - {self.email}"
````

````python
# 🔥 VERSION +++ (niveau encore au-dessus)
"""
On ajoute :
- encapsulation
- contrôle d’accès
"""

class Utilisateur:
    def __init__(self, nom, email):
        self.nom = nom
        self.email = email  # passe par le setter

    @property
    def email(self):
        return self._email

    @email.setter
    def email(self, valeur):
        if "@" not in valeur:
            raise ValueError("Email invalide")
        self._email = valeur

    def afficher_profil(self):
        return f"{self.nom} - {self.email}"
````
